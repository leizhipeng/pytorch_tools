# 1. Tensor Operations
* This notebook aims to collect codes for tensor operations.


```python
import torch
```

```python
# Create empty tensor.
x = torch.empty(5, 3)
print(x)
```

```python
# Create Random tensor.
x = torch.rand(5, 3)
print(x)
```

```python
# Create Zeros tensor.
x = torch.zeros(5, 3, dtype=torch.long)
print(x)
```

```python
# Create tensor from list.
x = torch.tensor([5.5, 3])
print(x)
```

```python
# Use existed tensor.
x = x.new_ones(5, 3, dtype=torch.double)      # new_*
print(x)

x = torch.randn_like(x, dtype=torch.float)
print(x)
```

```python
# size
print(x.size())
```

```python
# Add
y = torch.rand(5, 3)
print(x + y)
print(torch.add(x, y))
```

```python
# with _
y.add_(x)
print(y)
```

```python
# torch.view
x = torch.randn(4, 4)
y = x.view(16)
z = x.view(-1, 8)  #  size -1 deduced from other dim.
print(x.size(), y.size(), z.size())
```

```python
# Get value.
x = torch.randn(1)
print(x)
print(x.item())
```

```python
# numpy
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
np.add(a, 1, out=a)
print(a)
print(b)
```

```python
# CUDA
# is_available Check CUDA available
# ``torch.device`` more tensor to a device
if torch.cuda.is_available():
    device = torch.device("cuda")
    y = torch.ones_like(x, device=device)
    x = x.to(device)
    z = x + y
    print(z)
    print(z.to("cpu", torch.double))
```
